```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)
from mpl_toolkits.mplot3d import Axes3D
```

```python
data =pd.read_csv('student.csv')
print(data.shape)
data.head()
```

```python
math = data['Math'].values
read = data['Reading'].values
write = data['Writing'].values
```

```python
# Ploting the scores as scatter plot
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(math, read, write, color='#ef1234')
plt.show()
```

```python
# Create X, Y , W
m = len(math)
NUMBER_WEIGHTS = 3
# Vector de unos para agregar a la columna 0 de la matriz X, lo que representa la caracteristica independiente
x0 = np.ones(m)
#Creamos la matriz X de dimension 1000x3 con la primera columna en 1,seguidas de las calificaciones en math y read
#Esta se transpone debido a que al crearce se agregar primero el vector de 1 , luego math y read
X = np.array([x0,math,read]).T
# Inicializamos los coeficientes
#EL vector de pesos lo inicializamos en 0 y para este caso son 3 pesos w0, w1, w,2
W = np.array(np.zeros(NUMBER_WEIGHTS))
Y = np.array(write)
alpha = 0.0001

'X ',X.shape,'Y',Y.shape,'W',W.shape,X,W
```

```python
#Definir funcion de costo
def cost(X,Y,W):
    k = 1/(2*m)
    J = k*np.sum((X.dot(W)-Y)**2)
    return J
```

```python
inital_cost = cost(X, Y, W()
#print(inital_cost))
print(X.T)
```

```python
def gradient_descent(X, Y, W, alpha, iterations):
    cost_history = [0] * iterations
    m = len(Y)
    
    for iteration in range(iterations):
        # Hypothesis Values
        h = X.dot(W)
        # Difference b/w Hypothesis and Actual Y
        loss = h - Y
        # Gradient Calculation
        gradient = X.T.dot(loss) / m
        # Changing Values of B using Gradient
        W = W - alpha * gradient
        # New Cost Value
        _cost = cost(X, Y, W)
        cost_history[iteration] = _cost
        
    return W, cost_history
```

```python
# 100000 Iterations
newB, cost_history = gradient_descent(X, Y, W, alpha, 100000)

# New Values of B
print(newB)

# Final Cost of new B
print(cost_history[-1])
```
